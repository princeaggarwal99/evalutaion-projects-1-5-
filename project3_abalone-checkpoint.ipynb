```python
abalone_new=data[(z<3).all(axis=1)]
abalone_new.head()
```

```python
## data with removed outliers
abalone_new.shape
```

```python
# seperating target variables
x=abalone_new.iloc[:,0:-1]
y=abalone_new.iloc[:,-1]
```

```python
y.head()
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
x=scale.fit_transform(x)
```

```python
## scaled data with [0,1]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
## function for finding the optimum value for r_state giving best r2_Score:_


max_r_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.22)
    lreg=LinearRegression()
    lreg.fit(x_train,y_train)
    ylreg=lreg.predict(x_test)
    r2_scr=r2_score(y_test,ylreg)
    print('r2_score corr to random state',r_state,'is:',r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2 score corr to ",final_r_state,'is',max_r_score)       
        
```

```python
## 82 is the best random state for lr ,with 56.44 as the r2_score 
    
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=82,test_size=0.22)
```

```python
#random state max r2 score corr to  82 is 0.5644622397215956
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=82,test_size=0.22)
lr=LinearRegression()
lr.fit(x_train,y_train)
ylreg=lr.predict(x_test)
print(lr.score(x_test, y_test)*100)
print(mean_squared_error(ylreg, y_test))
print(mean_absolute_error(ylreg, y_test))
print(cross_val_score(lreg,x,y,cv=5,scoring="r2").mean())
print(r2_score(y_test,ylreg))
```

```python
# roc_lr
fpr_lr, tpr_lr, thresholds = roc_curve(y_test,ylreg,pos_label=2)
```

```python
## KNN_ROC
fpr_knn, tpr_knn, thresholds = roc_curve(y_test,knnpred,pos_label=2)
```

```python
### BEST RANDOM STATE FUNCTION FOR SVC
max_r_score=0
for rstate in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=rstate,test_size=0.22)
    svr=SVR(kernel='rbf')
    svr.fit(x_train,y_train)
    y_svr=svr.predict(x_test)
    r2scr=r2_score(y_test,y_svr)
    print('r2_score corr to random state',rstate,'is:',r2scr)
    if r2scr>max_r_score:
        max_r_score=r2scr
        final_r_state=rstate

print("max r2 score corr to ",final_r_state,'is',max_r_score)  
```

```python
#random state max r2 score corr to  45 is 0.578
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.22)
svr.fit(x_train,y_train)
y_svr=svr.predict(x_test)
print(svr.score(x_test, y_test)*100)
print(mean_squared_error(y_svr, y_test))
print(mean_absolute_error(y_svr, y_test))
print(cross_val_score(svr,x,y,cv=5,scoring="r2").mean())
print(r2_score(y_test,y_svr))

```

```python
max_r_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.22)
    gbr=GradientBoostingRegressor(learning_rate= 0.01, n_estimators=1000)
    gbr.fit(x_train,y_train)
    y_gbr=gbr.predict(x_test)
    r2=r2_score(y_test,y_gbr)
    print('r2_score corr to random state',r,'is:',r2)
    if r2>max_r_score:
        max_r_score=r2
        final_r_state=r

print("max r2 score corr to ",final_r_state,'is',max_r_score)  
```
