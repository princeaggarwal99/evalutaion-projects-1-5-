```python
## TREATMENT OF OULIERS
from scipy.stats import zscore
z_score=np.abs(zscore(sbi_data))
z_score
```

```python
threshold_value=3
print(np.where(z_score>3))
```

```python
z_score[71][3]
```

```python
## CORR HEATMAP 
sns.heatmap(sbi_data.corr(),annot=True)
```

```python
sbi_new=sbi_data[(z_score<3).all(axis=1)]
sbi_new
```

```python
sbi_new.shape
```

```python
## SEPERATION OF TARGET VARIABLE
x=sbi_new.iloc[:,0:-1]
y=sbi_new.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
```

```python
## STANDARD SCALING DATA BETWEEN [0,1]
x=SS.fit_transform(x)
x
```

```python
# SEPERATING TRAIN AND TEST DATA
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
## LOGISITC REGRESSION
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
ypred=lg.predict(x_test)
ypred
```

```python
## IMPORTING METRICS
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
accuracy_score(y_test,ypred)
```

```python
classification_report(y_test,ypred)
```

```python
confusion_matrix(y_test,ypred)
```

model

```python

from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

```

```python
## KNN CALSSIFIER
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
pred_knn=knn.predict(x_test)
print(accuracy_score(y_test,pred_knn))
print(confusion_matrix(y_test,pred_knn))
print(classification_report(y_test,pred_knn))
```

```python
svc=SVC()

```

```python
from sklearn.model_selection import GridSearchCV
## FOR FINDING THE OPTIMUM PARAMETERS
```

```python
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
CLF = GridSearchCV(svc, parameters)
CLF.fit(x,y)
print(CLF.best_params_)
```

```python
## SUPPORT VECTOR MACHINE
svc=SVC(kernel='rbf')
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
# DECISION TREE CLASSIFIER

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtcpred=dtc.predict(x_test)
print(accuracy_score(y_test,dtcpred))
print(confusion_matrix(y_test,dtcpred))
print(classification_report(y_test,dtcpred))
```

```python
## NAIVE BAYES = GAUSSIAN NB
gb=GaussianNB()
gb.fit(x_train,y_train)
gbpred=gb.predict(x_test)
print(accuracy_score(y_test,gbpred))
print(confusion_matrix(y_test,gbpred))
print(classification_report(y_test,gbpred))
```

```python
## RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
random_f=RandomForestClassifier()
```

```python
parameters = { 
    'n_estimators':[1,100,200,300,400,500,1000],
    'criterion' :['gini', 'entropy']
}
clf = GridSearchCV(random_f, parameters)
clf.fit(x,y)
print(clf.best_params_)
```

```python
random_f=RandomForestClassifier(n_estimators=300,criterion='entropy')
random_f.fit(x_train,y_train)
pred_rf=random_f.predict(x_test)
print(accuracy_score(y_test,pred_rf))
print(confusion_matrix(y_test,pred_rf))
print(classification_report(y_test,pred_rf))
```

```python
## ADA BOOST CALSSIFIER
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()
ada.fit(x_train,y_train)
ada_pred=ada.predict(x_test)
print(accuracy_score(y_test,ada_pred))
print(confusion_matrix(y_test,ada_pred))
print(classification_report(y_test,ada_pred))
```

```python
## at random_State=42  DECISION TREE CLASSIFIER GIVES THE BEST RESULT
```

```python
## DTC
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtcpred=dtc.predict(x_test)
print(accuracy_score(y_test,dtcpred)*100)
print(confusion_matrix(y_test,dtcpred))
print(classification_report(y_test,dtcpred))
```

```python
## CROSS VALIDATING THE SCORE OF DTC()

from sklearn.model_selection import cross_val_score
score=cross_val_score(dtc,x,y,cv=10)
print(score.mean()*100)
print(score.std()*100)
```
