```python
##  A=GOOD
    #   B=AVERAGE
    #C= BAD


grade=[]
for i in wine_data['quality']:
    if i>=1 and i <=3:
        grade.append('A')
    elif i>=4 and i<=7:
        grade.append('B')
    elif i>=7 and i<=10:
        grade.append('C')
wine_data['Grades']=grade
```

```python
wine_data.head()
```

```python
wine_data['Grades'].value_counts()
```

```python
sns.countplot(wine_data['Grades'])

## average counts are 1571 
## whereas the good wine is the lowest 10
## and bad quality is of 18 samples
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
##label encoding the grades so that work on an classification report
```

```python
wine_data['Grades']=le.fit_transform(wine_data['Grades'])
wine_data.head()
```

```python
sns.countplot(wine_data['Grades'])  ##B=1 AVERAGE  ,C=2 BAD, A=1 GOOD
              
```

```python
## seperating target and testing variables 
x = wine_data.iloc[:,:11]
y = wine_data['Grades']
```

```python
x.head(10)  ##first 10 rows of x (without traget variable)
```

```python
y.head() ## target column
```

```python
## scaling the data of ranging between 0-1 now
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x
```

```python
## finding the best correlated columns amongst 
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
x_pca = pca.fit_transform(x)
```

```python
print(x_pca)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
## spliting into train test data for testing the accuracy of model 
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,random_state=42,test_size=.22)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## LOGISTIC REGRESSION
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_lr = lr.predict(x_test)
print(confusion_matrix(y_test, y_lr))
print(accuracy_score(y_test, y_lr)*100)
```

```python
# roc_lr
fpr_lr, tpr_lr, thresholds = roc_curve(y_test,y_lr,pos_label=2)
```

```python
##DECISION TREE CLASSIFIER
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt_pred = dt.predict(x_test)
print(confusion_matrix(y_test, dt_pred))
print(accuracy_score(y_test, dt_pred))
```

```python
##DTR_ROC
fpr_dt, tpr_dt, thresholds = roc_curve(y_test,dt_pred,pos_label=2)
```

```python
## IMPORTING GRIDSEARCH FOR FINDING THE BEST PARAMETERS FOR THE MODEL
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
rfr=RandomForestClassifier()
```

```python
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(rfr,parameters ,scoring='accuracy', cv=10)
clfr.fit(x,y)
clfr.best_params_
```

```python
## RANDOM FOREST CLASSIFIER
rfr = RandomForestClassifier(n_estimators=10)
rfr.fit(x_train, y_train)
rfr_pred=rfr.predict(x_test)
print(confusion_matrix(y_test, rfr_pred))
print(accuracy_score(y_test, rfr_pred))
```

```python
##RANDOM_ROC
fpr_rfr, tpr_rfr, thresholds = roc_curve(y_test,rfr_pred,pos_label=2)
```

```python
## SVC SUPPORT VECTOR CLASSIFIER 
from sklearn.svm import SVC
svc=SVC()
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
grid_svc = GridSearchCV(svc, param_grid=param, scoring='accuracy', cv=10)
grid_svc.fit(x_train, y_train)
grid_svc.best_params_
```

```python
svc = SVC(C=1.2 ,kernel='rbf',gamma=1.4)
svc.fit(x_train, y_train)
svc_pred=svc.predict(x_test)
print(confusion_matrix(y_test, svc_pred))
print(accuracy_score(y_test, svc_pred))
```

```python
##SVR_ROC
fpr_svc, tpr_svc, thresholds = roc_curve(y_test,svc_pred,pos_label=2)
```

```python
## KNEIGHBOURS CALSSIFIER ()
knn=KNeighborsClassifier()
parameters={'n_neighbors':[i+1 for i in range(10)]}
clf=GridSearchCV(knn,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
print(accuracy_score(knn_pred,y_test))
classification_report(knn_pred,y_test)
```

```python
## KNN_ROC
fpr_knn, tpr_knn, thresholds = roc_curve(y_test,knn_pred,pos_label=2)
```

```python
##CROSS-VALIDATION SCORE OF RANDOM FOREST CLASSIFIER
score_random_forest=cross_val_score(rfr,x_train,y_train,cv=10)
score_random_forest.mean()
```

```python
##CROSS-VALIDATION SCORE OF LOGISTIC REGRESSION

score_lr=cross_val_score(lr,x_train,y_train,cv=10)
score_lr.mean()
```

```python
# random forest classifier is worling best witht the dataset with a accuracy score of 98.29

rfr = RandomForestClassifier(n_estimators=10)
rfr.fit(x_train, y_train)
rfr_pred=rfr.predict(x_test)
print(confusion_matrix(y_test, rfr_pred))
print(accuracy_score(y_test, rfr_pred))
```

```python
## cross val score of RANDOM FOREST ARE:-
score_random_forest=cross_val_score(rfr,x_train,y_train,cv=10)
print(score_random_forest.mean())
print(score_random_forest.std())
```
