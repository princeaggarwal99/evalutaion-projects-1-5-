```python
## dropping index variables and object variables from the dataset

bike=data.drop(['dteday', 'casual','yr', 'registered','instant'],axis=1)
bike
```

```python
## treating outliers
from scipy.stats import zscore
```

```python
z=np.abs(zscore(bike))
bike_data=bike[(z<3).all(axis=1)]
```

```python
x=bike_data.iloc[:,:-1]
y=bike_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import  mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingRegressor,AdaBoostRegressor,RandomForestRegressor
```

```python
## FUNCTION FOR FINDING BEST RANDOM STATE
```

```python
max_r_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.22)
    lreg=LinearRegression()
    lreg.fit(x_train,y_train)
    ylreg=lreg.predict(x_test)
    r2_scr=r2_score(y_test,ylreg)
    print('r2_score corr to random state',r_state,'is:',r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2 score corr to ",final_r_state,'is',max_r_score)       
        
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=51,test_size=0.22)
```

```python
#random state max r2 score corr to  51 is  0.3798492040297099
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=51,test_size=0.22)
lr=LinearRegression()
lr.fit(x_train,y_train)
ylreg=lr.predict(x_test)
print(lr.score(x_test, y_test)*100)
print(mean_squared_error(ylreg, y_test))
print(mean_absolute_error(ylreg, y_test))
print(cross_val_score(lreg,x,y,cv=5,scoring="r2").mean())
print(r2_score(y_test,ylreg))
```

```python
fpr_lr, tpr_lr, thresholds = roc_curve(y_test,ylreg,pos_label=2)
```

```python
fpr_knn, tpr_knn, thresholds = roc_curve(y_test,knnpred,pos_label=2)
```

```python
# FUNCTION FOR OPTIMUM VALUE OF R STATE
max_r_score=0
for rstate in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=rstate,test_size=0.22)
    svr=SVR(kernel='rbf')
    svr.fit(x_train,y_train)
    y_svr=svr.predict(x_test)
    r2scr=r2_score(y_test,y_svr)
    print('r2_score corr to random state',rstate,'is:',r2scr)
    if r2scr>max_r_score:
        max_r_score=r2scr
        final_r_state=rstate

print("max r2 score corr to ",final_r_state,'is',max_r_score)  
```

```python
fpr_svr, tpr_svr, thresholds = roc_curve(y_test,y_svr,pos_label=2)
```

```python
## DECISION TREE REGRESSOR
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtrpred=dtr.predict(x_test)

dtr.score(x_test,y_test)
print(r2_score(y_test,dtrpred))
```

```python
# CROSS_VALIDATION
score=cross_val_score(dtr,x,y,cv=10,scoring='r2')
score.mean()
```

```python
fpr_dtr, tpr_dtr, thresholds = roc_curve(y_test,dtrpred,pos_label=2)
```

```python
# RANDOM FORESST REGRESSOR
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(rfr,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
rfr = RandomForestRegressor(n_estimators = 400)
rfr.fit(x_train,y_train)
yrfr=rfr.predict(x_test)
print(rfr.score(x_test,y_test))
print(mean_squared_error(yrfr, y_test))
print(mean_absolute_error(yrfr, y_test))
print(r2_score(yrfr, y_test))
print(cross_val_score(rfr,x,y,cv=5,scoring='r2').mean())


```

```python
fpr_rfr, tpr_rfr, thresholds = roc_curve(y_test,yrfr,pos_label=2)
```

```python
## GRADIENT BOOSTING 
from sklearn.ensemble import GradientBoostingRegressor
gbr= GradientBoostingRegressor()
```

```python
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000],'learning_rate':[0.001,0.01,0.1,1]}
clfg=GridSearchCV(gbr,parameters)
clfg.fit(x,y)
clfg.best_params_
```

```python
max_r_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.22)
    gbr=GradientBoostingRegressor(learning_rate= 0.01, n_estimators=1000)
    gbr.fit(x_train,y_train)
    y_gbr=gbr.predict(x_test)
    r2=r2_score(y_test,y_gbr)
    print('r2_score corr to random state',r,'is:',r2)
    if r2>max_r_score:
        max_r_score=r2
        final_r_state=r

print("max r2 score corr to ",final_r_state,'is',max_r_score)  
```

```python
fpr_gbr, tpr_gbr, thresholds = roc_curve(y_test,y_gbr,pos_label=2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=51,test_size=0.22)
gbr=GradientBoostingRegressor(learning_rate= 0.1, n_estimators=200)
gbr.fit(x_train,y_train)
y_gbr=gbr.predict(x_test)
print(gbr.score(x_test, y_test)*100)
print(mean_squared_error(y_gbr, y_test))
print(mean_absolute_error(y_gbr, y_test))
print(cross_val_score(gbr,x,y,cv=5,scoring="r2").mean())
print(r2_score(y_test,y_gbr))
```

```python
### ADA BOSSTING
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
ada.fit(x_train,y_train)
yada=ada.predict(x_test)
print(ada.score(x_test,y_test))
print(mean_squared_error(yada, y_test))
print(mean_absolute_error(yada, y_test))
print(r2_score(yada, y_test))
print(cross_val_score(ada,x,y,cv=5,scoring='r2').mean())
```

```python
fpr_ada, tpr_ada, thresholds = roc_curve(y_test,yada,pos_label=2)
```
